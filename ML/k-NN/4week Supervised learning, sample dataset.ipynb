```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print(cancer.keys())
```

```python
print('유방암 데이터 형태: ', cancer['data'].shape)
```

```python
print('종양 종류', cancer['target_names'])
```

```python
print('종양', cancer['target'])
```

```python
print('클래스별 샘플 개수: \n', {n: v for n, v in zip(cancer['target_names'], np.bincount(cancer['target']))})
# n은 key, v는 value, zip은 쌍으로 합쳐주는 것. bincount는 각 target의 빈도수를 리턴해준다.
# 즉 각 target 이름에 따라 빈도수를 출력해준다.
```

```python
print('특성 이름:', cancer['feature_names'])
```
