# 2. 성과측정: 훈련데이터와 테스트 데이터

```python
from sklearn.model_selection import train_test_split

# train:test = 0.75 : 0.25

# feature and targer
# dataset을 split할 때 위처럼 target value가 정해져 있으면, split할 때, class 2가 훈련되지 못하는 경우도 있을 수 있다.
# 이를 방지하기 위해 train_test_split은 random하게 값들을 구하게 된다.
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print('X_train size :', X_train.shape)
print('y_train size :', y_train.shape)
```

```python
print('X_test size :', X_test.shape)
print('y_test size :', y_test.shape)
```

# 4. 첫번째 머신러닝 모델: K-NN 알고리즘

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
# 다른 머신러닝은 결과값이 나오는데, knn은 실제로 트레이닝이 일어나지는 않는다.
# 학습이 일어나지 않는 이유는?
# 다른 알고리즘들은 training을 통해 실제값과의 오차를 줄일 수 있는 최적화를 진행한다.
# 하지만 knn은 데이터들을 받아들여 cluster를 형성하기만 하면 그만이다.
# test값이 들어오면 그 값이 어느 값들과 제일 많이 가까운지를 찾는다. 
# knn의 학습 과정에서는 최적화가 존재하지 않는다. 그저 학습데이터를 배치하는 작업만 수행한다.
# 그렇기 때문에 ML 정의에 따른 Error를 통해 Performance를 계속해서 개선하는 '학습'이 일어나지 않는다.
```

```python
X_new = np.array([[5,2.9,1,0.2]])
print('X_new.shape:', X_new.shape)
```

```python
prediction = knn.predict(X_new)
print('pred:', prediction)
print('target name:', iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print('testset prediction:\n', y_pred)
```

```python
print('accuracy: {:,.2f}'.format(knn.score(X_test, y_test)))
```

```python
# 예측한 값과, 실제 정답
print('accuracy: {:,.2f}'.format(np.mean(y_pred == y_test)))
```

# 6. 요약

```python
# 이 네 줄로 모든게 요약된다.

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print('accuracy: {:,.2f}'.format(knn.score(X_test, y_test)))
```
